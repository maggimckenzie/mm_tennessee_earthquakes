```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
import seaborn as sns
```

```python
tn_quakes = pd.read_csv('../Data/tenn_eq.csv')
```

```python
tn_quakes['mag_round'] = tn_quakes['mag'].round(0)
```

```python
fig = px.histogram(tn_quakes, x="mag", color='mag_round', marginal="rug")
fig.show()
```

```python
f, ax = plt.subplots(figsize=(14, 10))
sns.despine(f)

sns.histplot(
    tn_quakes,
    x="mag", hue="mag_round",
#     multiple="stack",
    palette="Set2",
    edgecolor=".3",
    linewidth=.5,
#     log_scale=True,
    label={'mag':'Magnitude'}
)
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
# ax.set_xticks([500, 1000, 2000, 5000, 10000])
```

```python
mag_normalized = tn_quakes['mag_round'].value_counts(normalize=True, sort=False).sort_index()
```

```python
mag_normalized = mag_normalized*100
```

```python
fig = px.bar(mag_normalized,y='mag_round',
              color='mag_round',height=400,
            labels={'index':'Magnitude Base', 'mag_round': 'Percent of values'})
fig.show()
```
